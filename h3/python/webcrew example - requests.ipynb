本範例是熟悉網路資料抓取程式，會用到 Requests 套件

```python
import requests
from requests import get
import xml.etree.ElementTree as ET
```

讀取學校官網

 https://web.ckgsh.ntpc.edu.tw/nss/p/index


通常為了避免亂碼的問題，會加上 encoding 的設定

```python
web1 = requests.get('https://web.ckgsh.ntpc.edu.tw/nss/p/index')
web1.encoding = 'utf-8'
print(web1.text)
```

抓取上課講義

 https://sites.google.com/go.edu.tw/etuser-h3-it/python-%E7%A8%8B%E5%BC%8F/%E8%B3%87%E6%96%99%E6%8E%A2%E5%8B%98


```python
web2 = requests.get('https://sites.google.com/go.edu.tw/etuser-h3-it/python-%E7%A8%8B%E5%BC%8F/%E8%B3%87%E6%96%99%E6%8E%A2%E5%8B%98')
web2.encoding = 'utf-8'
print(web2.text)
```

抓取學校公告RSS

https://web.ckgsh.ntpc.edu.tw/nss/main/feeder/5a9759adef37531ea27bf1b0/iluV2Uy7463?f=normal&vector=private&static=false

```python
web3 = requests.get('https://web.ckgsh.ntpc.edu.tw/nss/main/feeder/5a9759adef37531ea27bf1b0/iluV2Uy7463?f=normal&vector=private&static=false')
web3.encoding = 'utf-8'
print(web3.text)
```

學校公告RSS是xml格式，所以要經過解析後才比較好使用

```python
tree = ET.ElementTree(ET.fromstring(web3.text))
root = tree.getroot()
print(root)
```

將公告中的 title 內容取出並列印出來

```python
for mydata in root.iter('title'):
    print(mydata.text)
```
